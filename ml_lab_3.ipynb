```python
#Mean squared error (cost function)
def compute_cost(y,y_pred):
  n=len(y)
  cost=(1/(2*n))*np.sum((y_pred-y))** 2
  return cost
```

```python
#Compute initial cost
cost=compute_cost(y,y_pred)
print("Initial Cost (MSE):",cost)
```

```python
#Given parameters
m=0.0
c=0.0
#prediction
y_pred=predict(x_norm,m,c)
#cost calculation
cost=compute_cost(y,y_pred)
print("cost for m=",m,"and c=",c,"is:",cost)
```

```python
learning_rate=0.01 #step size for gradient
epochs=100
print("learning rate:",learning_rate)
print("no of epochs:",epochs)
```

```python
n=len(y)
n
```

```python
cost_history=[]
```

```python
#Gradient Descent loop
for epoch in range(epochs):
  #step 1:prediction
  y_pred=predict(x_norm,m,c)
  #step 2:compute predictions
  dm=(1/n)*np.sum((y_pred-y)*x_norm)
  dc=(1/n)*np.sum((y_pred-y))
  #step 3:update parameters
  m=m-learning_rate*dm
  c=c-learning_rate*dc
  #step 4:Compute and store cost
  cost=compute_cost(y,y_pred)
  cost_history.append(cost)
  #print cost occasionally
  if epoch % 100==0:
    print(f"Epoch",epoch,"cost:",cost)
```
